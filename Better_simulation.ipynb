# Model rocket flight simulation
There are two scenarios explored by the simulation:
1) simulation for firing prograde motor only
2) simulation for firing both, prograde and retrograde motors

The simualtion can be executed for any rocket with calculated mass and acceleration (Thrust produced by the motor(s)). 
For the currrent simulation:
1) Mass of rocket = 1.505 kg
2) Thrust = 45 N
3) Therefore the acceleration expirienced by the rocket is 27.27 m/s^2
4) The burn time of each motor is 2 seconds

```python
# Import necessary libraries

import matplotlib.pyplot as plt
import math
%matplotlib inline
```

# Simulation for firing Prograde motor only
This simulation does not account a propulsive landing. Therefore the retromotor is not fired. The simulation helps in choosing a point to deploy parachute for safe recovery.

```python
time = [0, 0.25, 0.5, 0.75, 1, 1.25, 1.5, 1.75, 2, 2.25, 2.5, 2.75, 3, 3.25, 3.5, 3.75, 4, 4.25, 4.5, 4.75, 5, 5.25, 5.5, 5.75, 6, 6.25, 6.5, 6.75, 7, 7.25, 7.5, 7.75, 8, 8.25, 8.5, 8.75, 9, 9.1, 9.25, 9.5, 9.75, 10, 10.25, 10.5, 10.75, 11]
dist = []
vel = []
#u = 34.92
g = 9.81

for t in time:
    if t <=2:
        v = (27.27-g) * t
        s = (v*v) / (2*(27.27-g))
        dist.append(s)
        vel.append(v)
    else:
        v = vel[8] + ((-g)*(t-2))
        s = (vel[8]*(t-2)) + (0.5*(-g)*((t-2)*(t-2)))
        dist.append(s+34.92)
        vel.append(v)
        
print(dist)
print(vel) 

plt.plot(time, dist)
plt.plot(time, vel)
plt.grid(True)
plt.xlabel('Time')
plt.ylabel('Altitude and Velocity')
plt.title('Prograde only')
plt.legend(["Distance", "Velocity"])
```

# Simulation for firing both, prograde and retrograde motors
This simulation takes into account the propulsive landing. The retrograde motor is fired at a calculated time delay to aim for a soft touch-down.

```python
time = [0, 0.25, 0.5, 0.75, 1, 1.25, 1.5, 1.75, 2, 2.25, 2.5, 2.75, 3, 3.25, 3.5, 3.75, 4, 4.25, 4.5, 4.75, 5, 5.25, 5.5, 5.75, 6, 6.25, 6.5, 6.75, 7, 7.25, 7.5, 7.75, 8, 8.25, 8.5, 8.75, 9, 9.1, 9.25, 9.5, 9.75, 10, 10.25, 10.5, 10.75, 11, 11.25, 11.5, 11.75, 12, 12.25, 12.5, 12.75, 13, 13.25, 13.5, 13.75, 14]
dist = []
vel = []

g = 9.81
#u1 = 34.29
#u2 = -34.730999999999995

for t in time:
    if t <=2:
        v = (27.27-g) * t
        s = (v*v) / (2*(27.27-g))
        dist.append(s)
        vel.append(v)
    elif t>2 and t<=9.1:
        v = vel[8] + ((-g)*(t-2))
        s = (vel[8]*(t-2)) + (0.5*(-g)*((t-2)*(t-2)))
        dist.append(s+34.92)
        vel.append(v)
    elif t>9.1 and t<=11:
        v = vel[37] + ((27.27-9.81)*(t-9.1))
        s = (vel[37]*(t-9.1)) + (0.5*(27.27-g)*((t-9.1)*(t-9.1)))
        dist.append(s+34.92)
        vel.append(v)
    else:
        v = ((-g)*(t-11))
        s = (0.5*(-g)*((t-11)*(t-11)))
        dist.append(s)
        vel.append(v)
 
print(dist)
print(vel) 

plt.plot(time, dist)
plt.plot(time, vel)
plt.grid(True)
plt.xlabel('Time')
plt.ylabel('Altitude and Velocity')
plt.title('Prograde and retrograde')
plt.legend(["Distance", "Velocity"])
```
